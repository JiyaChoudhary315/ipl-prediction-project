```python
import pandas as pd

# Load the CSV files
matches = pd.read_csv("data/matches.csv")
deliveries = pd.read_csv("data/deliveries.csv")

# Show the first 5 rows of each
print("=== Matches Data ===")
display(matches.head())

print("=== Deliveries Data ===")
display(deliveries.head())
```

```python
print("Match columns:\n", matches.columns.tolist())
print("\nDeliveries columns:\n", deliveries.columns.tolist())

print("\nSeasons:\n", matches['season'].unique())
print("\nTeams:\n", matches['team1'].unique())
```

```python
print("Total matches:", matches.shape[0])
print("Winning teams:")
print(matches['winner'].value_counts())
```

```python
# Check for missing values in both datasets
print("Missing values in Matches Data:")
print(matches.isnull().sum())

print("\nMissing values in Deliveries Data:")
print(deliveries.isnull().sum())
```

```python
# Convert 'season' to integer
matches['season'] = matches['season'].astype(int)

# Convert 'date' column to datetime
matches['date'] = pd.to_datetime(matches['date'])
```

```python
# Check unique values in the 'season' column
print("Unique values in the 'season' column:")
print(matches['season'].unique())
```

```python
# Convert 'season' to the first year if it's a range
matches['season'] = matches['season'].apply(lambda x: str(x).split('/')[0] if isinstance(x, str) else x)

# Now try converting to integer
matches['season'] = matches['season'].astype(int)
```

```python
# Check unique values again to ensure proper conversion
print("Unique values in the 'season' column after conversion:")
print(matches['season'].unique())
```

```python
# Convert 'date' column to datetime format
matches['date'] = pd.to_datetime(matches['date'])

# Verify the change
print(matches['date'].head())
```

```python
# Check for missing values
print(matches.isnull().sum())

# Optionally, drop rows with missing values
matches = matches.dropna(subset=['winner'])

# Or, fill missing values (for example, with the most frequent value in 'venue')
matches['venue'] = matches['venue'].fillna(matches['venue'].mode()[0])
```

```python
import matplotlib.pyplot as plt

# Count the number of matches in each season
matches_per_season = matches['season'].value_counts().sort_index()

# Plot the number of matches per season
matches_per_season.plot(kind='bar', figsize=(10, 6))
plt.title("Number of Matches per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
```

```python
# Count the number of wins for each team
team_wins = matches['winner'].value_counts()

# Plot the number of wins for each team
team_wins.plot(kind='bar', figsize=(10, 6))
plt.title("Number of Wins by Each Team")
plt.xlabel("Team")
plt.ylabel("Number of Wins")
plt.show()
```

```python
# Count the most frequent venues
venue_counts = matches['venue'].value_counts()

# Display the top 10 venues
print(venue_counts.head(10))
```

```python
# Save the cleaned dataset to a new CSV file
matches.to_csv("ipl_matches_cleaned.csv", index=False)
```

```python
# Count the total number of wins for each team across all seasons
total_wins = matches['winner'].value_counts()

# Display the top 10 winning teams
print("Top 10 Winning Teams in IPL:")
print(total_wins.head(10))
```

```python
# Calculate the total runs scored by each batsman
player_runs = deliveries[deliveries['extra_type'].isnull()]  # Exclude extra runs
player_runs = player_runs.groupby('batsman')['total_runs'].sum().sort_values(ascending=False)

# Display the top 10 batsmen by total runs
print("Top 10 Batsmen by Total Runs:")
print(player_runs.head(10))
```

```python
# Exclude extra runs
player_runs = deliveries[deliveries['extra_runs'] == 0]  # Filter out extra runs
player_runs = player_runs.groupby('batter')['total_runs'].sum().sort_values(ascending=False)

# Display the top 10 batsmen by total runs
display(player_runs.head(10))
```

```python
# Calculate the total runs scored by each batsman
player_runs = deliveries[deliveries['extra_type'].isnull()]  # Exclude extra runs
player_runs = player_runs.groupby('batsman')['total_runs'].sum().sort_values(ascending=False)

# Display the top 10 batsmen by total runs
print("Top 10 Batsmen by Total Runs:")
print(player_runs.head(10))
```

```python
# Calculate the total wickets taken by each bowler
player_wickets = deliveries[deliveries['dismissal_kind'].notnull()]
player_wickets = player_wickets.groupby('bowler')['dismissal_kind'].count().sort_values(ascending=False)

# Display the top 10 bowlers by total wickets
print("Top 10 Bowlers by Total Wickets:")
print(player_wickets.head(10))
```
